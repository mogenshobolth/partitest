This notebook contains explorative data analysis (EDA) of the roll call vote data.

```python
import pandas as pd
df = pd.read_parquet('data.parquet')
```

```python
df.describe()
```

```python
df[["ministeromraade"]].value_counts()
```

```python
corr = df[['stem_FG','stem_ALT','stem_EL','stem_SF','stem_RV','stem_S','stem_V','stem_KF','stem_KD','stem_LA','stem_DF','stem_NB']]
corr.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.subplots(figsize=(10,10))
ax = sns.heatmap(
    corr.corr(), 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(240, 10, n=9),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
